```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('brain_2d_seg_data/training/brain_masks/100206_brain.png')
img2 = cv2.imread('brain_2d_seg_data/training/ventricle_masks/100206_ventricles.png')
```

```python
# show all data
np.set_printoptions(threshold=np.inf) 
```

# 合成一张label

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
gray1=gray1/255
gray2=gray2/255*2
```

```python
label = gray1+gray2  #重叠
plt.imshow(label)
```

```python
np.unique(label)
```

```python
label[label==3]=2  #把3都替换成2
print(label)
```

```python
np.unique(label)
```

```python
label.shape
```

# 保存labels

## matplotlib.image.imsave

```python
import os
SavePath_mat = "./brain_2d_seg_data/training/labels_mat/"
if not os.path.exists(SavePath_mat):
    os.mkdir(SavePath_mat)
    
matplotlib.image.imsave(SavePath_mat + "100206_labels.png", label)

# save 之后为rgb 3 通道png
```

## cv2.imwrite

```python
import os
SavePath_cv = "./brain_2d_seg_data/training/labels_cv/"
if not os.path.exists(SavePath_cv):
    os.mkdir(SavePath_cv)

SavePath_cv = "./brain_2d_seg_data/training/labels_cv/"
cv2.imwrite(SavePath_cv + "100206_labels.png",label)
```
