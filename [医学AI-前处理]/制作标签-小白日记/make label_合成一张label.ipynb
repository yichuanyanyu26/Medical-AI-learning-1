```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('brain_2d_seg_data/training/brain_masks/100206_brain.png')
img2 = cv2.imread('brain_2d_seg_data/training/ventricle_masks/100206_ventricles.png')
```

```python
# show all data
np.set_printoptions(threshold=np.inf) 
```

# 合成一张label

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
gray1.shape
```

```python
gray1.dtype
```

```python
gray2
```

```python
gray1 = gray1 / 255
gray2 = gray2 / 255 * 2
```

```python
label_col = np.vstack((gray1, gray2))               #纵向连接
label_row = np.concatenate([gray1, gray2], axis=1)  #横向连接
```

```python
plt.imshow(label_col)
```

```python
plt.imshow(label_row)
```

```python
label = gray1+gray2  #重叠
plt.imshow(label)
```
