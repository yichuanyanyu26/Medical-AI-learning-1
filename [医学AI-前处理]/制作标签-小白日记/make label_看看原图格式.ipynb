### cv2.imshow

```python
cv2.imshow("img", img1[:,:,2])
cv2.waitKey(0)
```

### cv2 没有subplot

```python
cv2.imshow("img",img1[:,:,0])
cv2.waitKey(0)
cv2.imshow("img",img1[:,:,1])
cv2.waitKey(0)
cv2.imshow("img",img1[:,:,2])
cv2.waitKey(0)
```

```python
label_col = np.vstack((img1[:,:,0], img1[:,:,1],img1[:,:,2]))               #纵向连接
cv2.imshow("img",label_col)
cv2.waitKey(0)
```

```python
label_row = np.concatenate([img1[:,:,0], img1[:,:,1],img1[:,:,2]], axis=1)  #横向连接
cv2.imshow("img",label_row)
cv2.waitKey(0)
```

### plt.imshow

```python
import matplotlib.pyplot as plt
plt.imshow(img1)
```

```python
rgb_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img[:,:,2])
```

```python
plt.figure()
plt.subplot(1,3,1)
plt.imshow(img1[:,:,0])
plt.subplot(1,3,2)
plt.imshow(img1[:,:,1])
plt.subplot(1,3,3)
plt.imshow(img1[:,:,2])
```

```python
plt.figure()
plt.subplot(1,3,1)
plt.imshow(img1[:,:,0],cmap="gray")
plt.subplot(1,3,2)
plt.imshow(img1[:,:,1],cmap="gray")
plt.subplot(1,3,3)
plt.imshow(img1[:,:,2],cmap="gray")
```
